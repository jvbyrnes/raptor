```python
from pymongo import MongoClient
from dotenv import load_dotenv
import os
import certifi
from raptor import RetrievalAugmentation
import logging

logging.basicConfig(level=logging.DEBUG)

load_dotenv()
```

```python
client = MongoClient(os.environ.get("MONGO_URI"),tlsCAFile=certifi.where())

knowledge_hub_articles = client['corpus']['knowledge_hub_approved_clean_parsed'].find(limit=10)

knowledge_hub_text = ''.join([article['content'] for article in knowledge_hub_articles])

print(knowledge_hub_text)
```

```python
RA = RetrievalAugmentation()

RA.add_documents(knowledge_hub_text)
```

```python
question = "what has ooho done for plastics use?"

answer = RA.answer_question(question)

print("Answer: ", answer)
```

```python
question = "What does recognitionAI do?"

answer = RA.answer_question(question)

print("Answer: ", answer)
```
